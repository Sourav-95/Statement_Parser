```python
import PyPDF2
import camelot

def unlock_pdf(input_pdf, output_pdf, password):
    pdf_reader = PyPDF2.PdfReader(input_pdf)
    if pdf_reader.is_encrypted:
        pdf_reader.decrypt(password)
        pdf_writer = PyPDF2.PdfWriter()

        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            pdf_writer.add_page(page)

        with open(output_pdf, 'wb') as f:
            pdf_writer.write(f)
        print(f"Unlocked PDF saved as {output_pdf}")
    else:
        print("The PDF is not encrypted.")

def extract_tables_from_pdf(pdf_path):
    tables = camelot.read_pdf(pdf_path)
    return tables

# Usage
input_pdf = '/Users/souravm/Documents/Projects/statement_parser/statement_20250404.pdf'
output_pdf = '/Users/souravm/Documents/Projects/statement_parser/statement_2025040_un.pdf'
password = 'SOUR844767170'

unlock_pdf(input_pdf, output_pdf, password)
tables = extract_tables_from_pdf(output_pdf)

# Display the extracted tables
# for idx, table in enumerate(tables):
#     print(f"Table {idx + 1}")
#     print(table)
```
